```python
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import random

# ======================================================
# CONFIG
# ======================================================

IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 70
NUM_CLASSES = 3
weight_decay = 1e-4

train_dir = "/kaggle/input/datasets/hitakshigabani/dataset-tb-and-pneumoniadetection/chest-xray-dataset/train_balanced"
val_dir = "/kaggle/input/datasets/hitakshigabani/dataset-tb-and-pneumoniadetection/chest-xray-dataset/val"
test_dir = "/kaggle/input/datasets/hitakshigabani/dataset-tb-and-pneumoniadetection/chest-xray-dataset/test"

output_dir = "/kaggle/working"
os.makedirs(output_dir, exist_ok=True)

# ======================================================
# DATA GENERATORS
# ======================================================

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=10,
    zoom_range=0.1,
    horizontal_flip=True
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

val_gen = val_test_datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'
)

test_gen = val_test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=1,
    class_mode='categorical',
    shuffle=False
)

# ======================================================
# RESNET BLOCKS
# ======================================================

def identity_block(x, filters):
    f1, f2, f3 = filters

    shortcut = x

    x = layers.Conv2D(f1, (1,1), kernel_regularizer=regularizers.l2(weight_decay))(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Conv2D(f2, (3,3), padding='same',
                      kernel_regularizer=regularizers.l2(weight_decay))(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Conv2D(f3, (1,1),
                      kernel_regularizer=regularizers.l2(weight_decay))(x)
    x = layers.BatchNormalization()(x)

    x = layers.Add()([x, shortcut])
    x = layers.ReLU()(x)

    return x


def conv_block(x, filters, stride=2):
    f1, f2, f3 = filters

    shortcut = layers.Conv2D(f3, (1,1), strides=stride,
                             kernel_regularizer=regularizers.l2(weight_decay))(x)
    shortcut = layers.BatchNormalization()(shortcut)

    x = layers.Conv2D(f1, (1,1), strides=stride,
                      kernel_regularizer=regularizers.l2(weight_decay))(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Conv2D(f2, (3,3), padding='same',
                      kernel_regularizer=regularizers.l2(weight_decay))(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    x = layers.Conv2D(f3, (1,1),
                      kernel_regularizer=regularizers.l2(weight_decay))(x)
    x = layers.BatchNormalization()(x)

    x = layers.Add()([x, shortcut])
    x = layers.ReLU()(x)

    return x

# ======================================================
# BUILD RESNET50
# ======================================================

def build_resnet50(input_shape=(224,224,3), classes=3):

    inputs = layers.Input(shape=input_shape)

    x = layers.ZeroPadding2D((3,3))(inputs)
    x = layers.Conv2D(64, (7,7), strides=(2,2),
                      kernel_regularizer=regularizers.l2(weight_decay))(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.MaxPooling2D((3,3), strides=(2,2), padding='same')(x)

    # Stage 1
    x = conv_block(x, [64,64,256], stride=1)
    for _ in range(2):
        x = identity_block(x, [64,64,256])

    # Stage 2
    x = conv_block(x, [128,128,512])
    for _ in range(3):
        x = identity_block(x, [128,128,512])

    # Stage 3
    x = conv_block(x, [256,256,1024])
    for _ in range(5):
        x = identity_block(x, [256,256,1024])

    # Stage 4
    x = conv_block(x, [512,512,2048])
    for _ in range(2):
        x = identity_block(x, [512,512,2048])

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.4)(x)

    outputs = layers.Dense(classes,
                           activation='softmax',
                           kernel_regularizer=regularizers.l2(weight_decay))(x)

    return models.Model(inputs, outputs)

model = build_resnet50(classes=NUM_CLASSES)

# ======================================================
# CLASS WEIGHTS
# ======================================================

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_gen.classes),
    y=train_gen.classes
)

class_weights = dict(enumerate(class_weights))

# ======================================================
# COSINE LR
# ======================================================

lr_schedule = tf.keras.optimizers.schedules.CosineDecay(
    initial_learning_rate=5e-5,
    decay_steps=EPOCHS * len(train_gen)
)

optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)

model.compile(
    optimizer=optimizer,
    loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
    metrics=['accuracy']
)

# ======================================================
# SAVE BEST MODEL ONLY
# ======================================================

checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath=os.path.join(output_dir, "resnet50_best.h5"),
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

# ======================================================
# TRAIN
# ======================================================

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    class_weight=class_weights,
    callbacks=[checkpoint],
    verbose=1
)

# ======================================================
# LOAD BEST MODEL
# ======================================================

best_model = tf.keras.models.load_model(
    os.path.join(output_dir, "resnet50_best.h5")
)

# ======================================================
# TEST EVALUATION
# ======================================================

loss, accuracy = best_model.evaluate(test_gen)
print("\nFinal Test Accuracy:", accuracy)

preds = best_model.predict(test_gen)
y_pred = np.argmax(preds, axis=1)
y_true = test_gen.classes

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred,
      target_names=test_gen.class_indices.keys()))

# ======================================================
# CONFUSION MATRIX
# ======================================================

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=test_gen.class_indices.keys(),
            yticklabels=test_gen.class_indices.keys())
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("ResNet50 Confusion Matrix")
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# ==============================
# CONFIG
# ==============================

IMG_SIZE = 224
model_path = "/kaggle/working/resnet50_best.h5"
test_dir = "/kaggle/input/datasets/hitakshigabani/dataset-tb-and-pneumoniadetection/chest-xray-dataset/test"

# ==============================
# LOAD MODEL
# ==============================

model = tf.keras.models.load_model(model_path)
print("Model loaded successfully.")

# ==============================
# TEST GENERATOR
# ==============================

test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)

test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=1,
    class_mode='categorical',
    shuffle=False
)

# ==============================
# EVALUATE
# ==============================

loss, accuracy = model.evaluate(test_gen, verbose=1)

print("\n==============================")
print("FINAL TEST RESULTS")
print("==============================")
print(f"Final Test Accuracy: {accuracy*100:.2f}%")
print(f"Final Test Loss: {loss:.4f}")

# ==============================
# PREDICTIONS
# ==============================

preds = model.predict(test_gen)
y_pred = np.argmax(preds, axis=1)
y_true = test_gen.classes

print("\nClassification Report:\n")
print(classification_report(
    y_true,
    y_pred,
    target_names=test_gen.class_indices.keys()
))

# ==============================
# CONFUSION MATRIX
# ==============================

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=test_gen.class_indices.keys(),
            yticklabels=test_gen.class_indices.keys())
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("ResNet50 Confusion Matrix")
plt.show()

# ==============================
# TEST RANDOM IMAGES
# ==============================

class_labels = list(test_gen.class_indices.keys())

for i in random.sample(range(len(test_gen.filepaths)), 5):
    img_path = test_gen.filepaths[i]
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = tf.keras.preprocessing.image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)
    predicted_class = class_labels[np.argmax(prediction)]

    plt.imshow(img)
    plt.title(f"Predicted: {predicted_class}")
    plt.axis("off")
    plt.show()
```
